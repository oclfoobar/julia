```julia
function simple_diff_quotient(xs, ys)
    length(xs) != length(ys) && error("E: simple_diff_quotient: wrong argument sizes")
    quotients = []
    
    for i in 2:length(xs)
        x = xs[i] - xs[i-1]
        x == 0 && error("E: simple_diff_quotient: wrong x value at index $i")
        push!(quotients, (ys[i]-ys[i-1])/x)
    end

    return quotients
end
```

Um die verrauschten Daten doch noch analysieren zu können, erinnern wir uns daran, dass die Ableitung einer Funktion letztendlich nichts anderes als eine lokale lineare Approximation ist. Wir können also die Ableitung berechnen, indem wir an jeden Punkt eine Tangente anlegen. Diese berechnen wir mittels linearer Regression. Die Tangente an Punkt $x_i$ ist gegeben durch
$y=b_i \cdot x + a_i,$
wobei hier die Steigung $b_i$ schon die Ableitung am Punkt $x_i$ ist. Diese Steigung $b_i$ ergibt sich aus:

$b_i = \frac{\sum_{j=i-n}^{i+n}\left(x_j-\bar{x}_i\right)\left(y_j-\bar{y}_i\right)}{\sum_{j=i-n}^{i+n}\left(x_j - \bar{x}_i\right)^2}.$

Die Mittelwerte sind definiert durch

$\bar{x}_i = \frac{1}{2n+1}\sum\limits_{j=i-n}^{i+n} x_j, \qquad \bar{y}_i = \frac{1}{2n+1}\sum\limits_{j=i-n}^{i+n} y_j.$

Berechnen Sie die Ableitung nun dreimal für $n={5, 10, 20}$ und plotten Sie alles in einem Plot. Welche Probleme können auftreten, wenn man $n$ zu groß wählt?

```julia
# x,y Daten als Argumente wie zuvor, nun zusaetzlich die Ordnung n
function difference_quotient(xs, ys, n)
    x_values  = []
    quotients = []
    
    for i in n+1:length(xs)-(n+1)
        # get mean value
        x_bar = sum(collect(xs[i-n:i+n]))/(2n+1)
        y_bar = sum(collect(ys[i-n:i+n]))/(2n+1)
        
        # sum up numerator and denominator
        num = 0
        den = 0
        for j in i-n:i+n
            num = (xs[j]-x_bar)*(ys[j]-y_bar)
            den = (xs[j]-x_bar)^2
        end
        
        push!(quotients, num/den)
        push!(x_values, xs[i])
    end

    return (x_values, quotients)
end
```

```julia
figure()
for n in [5, 10, 20]
    values = difference_quotient(x_values_2, y_values_2, n)
    plot(values[1], values[2], label="n = $n")
end
legend()
;
```

Bei dieser Art der Berechnung sollte man ein möglichst geringes $n$ wählen, damit man möglichst viele Werte zum Plotten hat. Man kann in dem Plot sehen, dass der Graph für $n=20$ ein geringeres Rauschen hat als für $n=5$, aber auch, dass er einen geringeren Bereich auf der x-Achse abdeckt.  
Dies liegt daran, dass auf beiden Seiten des Intervalls, in dem Werte vorliegen, $n+1$ Werte weniger dargestellt werden können, da ansonsten die Summe $\sum_{j=i-n}^{i+n}(\dots)$ nicht berechnet werden kann. Somit fehlen insgesamt $2n+2$ Werte. Bei hohem $n$ und wenigen Werten kann somit die Brauchbarkeit der Ableitung deutlich reduziert sein.
