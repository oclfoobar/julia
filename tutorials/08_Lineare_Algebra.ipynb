## 2. Lineare Gleichungssyteme

Betrachten wir ein lineares Gleichungssystem mit drei Unbekannten, etwa
\begin{eqnarray}
x & - 2y & +2z & = & 5\\
x & - y & +2z & = & 7 \\
-x & + y & + z & = & 5
\end{eqnarray}

```julia
A = [1 -2 2; 1 -1 2; -1 1 1]
```

```julia
b = [5, 7, 5]
```

Damit haben wir ein Gleichungssystem der Form
\begin{equation}
      A \cdot \vec{x} = \vec{b} \,,
\end{equation}
welches wir nach dem unbekannten Vektor $\vec{x}$ auflösen wollen.

```julia
x = A\b
```

```julia
A * x
```

```julia
using PyPlot

Einstein = imread("http://www.thp.uni-koeln.de/trebst/Lectures/CompPhys-2016/einstein.png")
imshow(Einstein, cmap="gray")
axis("off")
```

```julia
Einstein
```

```julia
λ_einstein = eigvals(Einstein)
```

```julia
plot(λ_einstein)
xlim(-5, 485)
```

```julia
S_einstein = svdvals(Einstein)
```

```julia
plot(S_einstein, "o")
xlim(-5, 485)
```

```julia
loglog(S_einstein, "o")
```

```julia
Zufall = rand(480,480);
```

```julia
S_zufall = svdvals(Zufall)
```

```julia
loglog(S_zufall, "o")
```

```julia
loglog(S_einstein, "o", label="Einstein")
loglog(S_zufall, "o", label="Zufall")
legend()
```
